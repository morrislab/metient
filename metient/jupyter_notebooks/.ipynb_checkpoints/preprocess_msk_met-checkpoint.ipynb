## Use MSK-MET data to get frequency of primary -> metastatic site migrations

```python
import os
import csv
repo_dir = "/Users/divyakoyyalagunta/Desktop/Cornell_Research/Morris_Lab/met_history_prediction/"
os.chdir(repo_dir)
data_dir = os.path.join(repo_dir, "src/data/msk_met")
csv_fn = os.path.join(data_dir, "msk_met_2021_clinical_data.tsv")
```

```python
import pandas as pd
df = pd.read_csv(csv_fn, sep = '\t')
df
```

```python
df[df['Sample Type']=='Metastasis']["Patient ID"]
```

```python
cols_to_keep = ["Primary Tumor Site"] + [x for x in list(df.columns) if "Distant Mets:" in x]
df = df[cols_to_keep]
df = df.replace(to_replace = ['Yes','No'],value = [1,0])
# Only keep samples with one met
df = df[df.sum(axis=1) == 1]
df = df.groupby("Primary Tumor Site").sum()
df.columns = [x.replace("Distant Mets: ", "") for x in list(df.columns)]
df
```

```python
row_sums = df.sum(axis=1)
print(row_sums)
freq = df.div(row_sums, axis=0)
freq
```

```python
list(df.index)
```

```python
freq.to_csv(os.path.join(data_dir, "msk_met_freq.csv"))
```

```python
freq_df = pd.read_csv(os.path.join(data_dir, "msk_met_freq.csv"))
freq_df
```

```python
import numpy as np
hoadley_to_msk_met_map = {
    "liver": "Liver",
    "brain": "CNS/Brain",
    "rib": "Bone",
    "breast": "Breast",
    "kidney": "Kidney",
    "lung": "Lung"
}
ordered_sites = ['breast', 'rib', 'lung', 'liver', 'brain']

organotrop_mat = np.zeros((len(ordered_sites), len(ordered_sites)))

mapped_sites = [hoadley_to_msk_met_map[site] for site in ordered_sites]
for i, start in enumerate(mapped_sites):
    for j, dest in enumerate(mapped_sites):
        print(i, j, start, dest)
        missing_val = False
        if start not in list(freq_df['Primary Tumor Site']):
            print(f"{start} not in MSK-MET as primary tumor")
            missing_val = True
        if dest not in list(freq.columns):
            print(f"{dest} not in MSK-MET as metastatic site")
            missing_val = True
        if missing_val:
            organotrop_mat[i,j] = 0.0 # TODO: what do we do when primary or met are not in db
        else:
            organotrop_mat[i,j] = freq_df[freq_df['Primary Tumor Site']==start][dest].item()

organotrop_mat
```
