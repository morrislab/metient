### Load paths

```python
import pyreadr
import os
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

REPO_DIR = os.path.join(os.getcwd(), "../../")
os.chdir(REPO_DIR)

TRACERX_DATA_DIR = os.path.join(REPO_DIR, "src/data/tracerx_nsclc/")
OUTPUT_DIR = os.path.join(REPO_DIR, "src/data/tracerx_nsclc/patient_tsvs")
```

### Load sample overview information (from TRACERx publicly available data)

```python
sample_info_df= pd.read_csv(os.path.join(TRACERX_DATA_DIR,"sample_overview_original.txt"), sep="\t")
print(sample_info_df['sampleType'].value_counts())
print(sample_info_df['sampleTypeDetail'].value_counts())
sample_info_df
```

### Load purity and ploidy information (from TRACERx publicly available data)

```python
purity_ploidy_df = pd.read_csv(os.path.join(TRACERX_DATA_DIR, "20220808_purityandploidy.txt"), sep="\t")
purity_ploidy_df
```

### Load all patient mutation data (from TRACERx publicly available data)

```python
patient_data_all = pyreadr.read_r(os.path.join(TRACERX_DATA_DIR, 'mutTableAll.cloneInfo.20220726.rda'))
patient_data_all_df = patient_data_all['mutTableAll']
print("Patient data columns:\n ", patient_data_all_df.columns)
print(patient_data_all_df[:10]['RegionSum'])

patient_data_all_df = patient_data_all_df[['mutation_id', 'patient_id', 'Hugo_Symbol','exonic.func','NucleotideChange', 'RegionSum' , 'MutCPN_SC']]
unique_patients = patient_data_all_df['patient_id'].unique()
print("Number of unique patients:", len(unique_patients))
patient_data_all_df
```

### Update sample overview df since some sample information is missing 

```python
'''
Examples of missing sample info:
CRUK0361_SU_FLN1
CRUK0702_SU_LN2
CRUK0027_SU_T1.R3
'''

rows_to_add = []
missing_sample_names = []
for patient_id in unique_patients:
    sample_names = patient_data_all_df[patient_data_all_df['patient_id']==patient_id]['RegionSum'].iloc[0]
    sample_names = [patient_id +"_"+x.split(":")[0] for x in sample_names.split(";")]
    for sample_name in sample_names:
        sample_info = sample_info_df[sample_info_df['region']==sample_name]
        if len(sample_info) == 0:
            missing_sample_names.append(sample_name)
            # Use the same sampleType and sampleTypeDetail as the other matching samples for that patient
            # e.g. CRUK0361_SU_FLN1 and CRUK0361_SU_FLN2 should get the sample sample info
            patient_samples = sample_info_df[sample_info_df['patient_id']==patient_id]
            if ("FLN" in sample_name) or ("LN" in sample_name):
                rows_to_add.append([patient_id, patient_id, sample_name, "metastasis", "LN"])
            else:
                matching = patient_samples[patient_samples['region'].str.contains(sample_name.split(".")[0])].iloc[0]
                rows_to_add.append([patient_id, matching['tumour_id'], sample_name, matching['sampleType'], matching['sampleTypeDetail']])

print(missing_sample_names)
updated_sample_df = pd.concat([sample_info_df, pd.DataFrame(rows_to_add, columns=sample_info_df.columns)], ignore_index=True)
updated_sample_df.to_csv(os.path.join(TRACERX_DATA_DIR,"sample_overview_complete.txt"), index=False, sep='\t')
```

### Prep data for input into Metient and MACHINA


```python
'''
Make a tsv with minimum columns: ['#sample_index', 'sample_label', 'anatomical_site_index',
'anatomical_site_label', 'character_index', 'character_label','ref', 'var']
'''

from src.util import pairtree_data_extraction_util as pt_util

def generate_sample_and_mut_info(row):
    '''
    Given a row in the patient_data_all_df dataframe, yields a 
    list with [sample_name, mutation_name, sample_type_detailed, ref_count,
     var_count, sample_type, purity, ploidy]
    e.g. ["SETD3:2:14:99872776:T", "BR_LN1", "LN_SU_FLN1", 43,270, "metastasis", 0.54, 3.19]
    '''
    mut_id = ":".join(row['mutation_id'].split(":")[1:])
    mut_name = f"{row['Hugo_Symbol']}:{mut_id}"
        
    # RegionSum and MutCPN_SC column values need to be split by sample 
    # (e.g. "SU_FLN1:0/562;SU_LN1:3/616")
    region_sums = row['RegionSum'].split(";")
#     if type(row['MutCPN_SC']) != str:
#         region_copy_numbers = [float('nan') for _ in range(len(region_sums))]
#     else:
#         region_copy_numbers = row['MutCPN_SC'].split(";")
        
    for mut_count in region_sums:
        # Lookup sample info in sample_overview dataframe
        sample_name = row['patient_id']+"_"+mut_count.split(":")[0]
        sample_info = updated_sample_df[updated_sample_df['region']==sample_name]
        if len(sample_info) == 0:
            yield None
            return
        
        sample_type = sample_info['sampleType'].item()
        sample_type_detailed = sample_info['sampleTypeDetail'].item()+"_"+mut_count.split(":")[0]
        
        # Lookup purity and ploidy info
        purity_ploidy_info = purity_ploidy_df[purity_ploidy_df["Sample"]==sample_name]
        if len(purity_ploidy_info) == 0:
            purity, ploidy = float("nan"), float("nan")
        else:
            purity = purity_ploidy_info["final.purity"].item()
            ploidy =  purity_ploidy_info["final.ploidy"].item()
        
        # Get ref and var counts, # TODO confirm with TRACERx that these are the right values
        counts = mut_count.split(":")[1] # e.g. "40/168"
        var = int(counts.split("/")[0])
        total = int(counts.split("/")[1])
        ref = total - var
        
        yield [mut_name, sample_name, sample_type_detailed, ref, var, sample_type, purity, ploidy]

for patient_id in unique_patients:
    print(patient_id)
    output_rows = []
    subset = patient_data_all_df[patient_data_all_df['patient_id'] == patient_id]
    print(len(subset))
    # e.g. [["BR_LN1", 43,270,3.3], ["BR_LN2",64,110,2.09]]
    sample_mut_info = []
    missing_info = False
    for _, row in subset.iterrows(): # each row is one mutation
        generator = generate_sample_and_mut_info(row)
        for data in generator:
            if data == None:
                missing_info = True
                break
            sample_mut_info.append(data)
                
    if missing_info:
        print("Missing sample info, skipping patient", patient_id)
        continue
        
    sample_mut_df = pd.DataFrame(sample_mut_info)
    sample_mut_df.columns = ["character_label", "sample_label", "anatomical_site_label", "ref",
                            "var", "sample_type", "purity", "ploidy"]
    
    assert(len(sample_mut_df['anatomical_site_label'].unique()) == len(sample_mut_df['sample_label'].unique()))
    # Add indices for mutations, samples and anatomical sites as needed for input format
    sample_mut_df['#sample_index'] = sample_mut_df.apply(lambda row: list(sample_mut_df['sample_label'].unique()).index(row["sample_label"]), axis=1)    
    sample_mut_df['character_index'] = sample_mut_df.apply(lambda row: list(sample_mut_df['character_label'].unique()).index(row["character_label"]), axis=1)
    sample_mut_df['anatomical_site_index'] = sample_mut_df.apply(lambda row: list(sample_mut_df['anatomical_site_label'].unique()).index(row["anatomical_site_label"]), axis=1)
    
    sample_mut_df = sample_mut_df[['#sample_index', "sample_label", "anatomical_site_index", "anatomical_site_label", 
                                 'character_index', "character_label", "ref","var", "sample_type","purity", "ploidy"]]

    sample_mut_df.to_csv(os.path.join(OUTPUT_DIR, f"{patient_id}_SNVs.tsv"), sep="\t")
    pt_util.write_pairtree_inputs(sample_mut_df, patient_id, OUTPUT_DIR)
```
