```python
import pandas as pd
import os

data_dir = "/Users/divyakoyyalagunta/Desktop/Cornell_Research/Morris_Lab/met_history_prediction/src/data/"
os.chdir(data_dir)


# Breast cancers
all_breast_cancer_studies = pd.read_csv(os.path.join(data_dir, 'cbio_all_breast_cancer_studies.csv'))
breast_sample_types = all_breast_cancer_studies["Sample Type"].unique()
print(f"Breast sample types: {breast_sample_types}")
breast_filtered_df = all_breast_cancer_studies.groupby('Patient ID').filter(lambda x: {'Primary', 'Metastasis'}.issubset(set(x['Sample Type'])))
num_breast_patients = len(breast_filtered_df["Patient ID"].unique())
print(f"Number of Breast cancer patients: {num_breast_patients}")

# Colorectal cancers
print()
all_crc_cancer_studies = pd.read_csv(os.path.join(data_dir, 'cbio_all_colorectal_cancer_studies.csv'))
crc_sample_types = all_crc_cancer_studies["Sample Type"].unique()
print(f"CRC sample types: {crc_sample_types}")
crc_filtered_df = all_crc_cancer_studies.groupby('Patient ID').filter(lambda x: {'Primary', 'Metastasis'}.issubset(set(x['Sample Type'])) or {'Primary Tumor', 'Metastasis'}.issubset(set(x['Sample Type'])))
num_crc_patients = len(crc_filtered_df["Patient ID"].unique())
print(f"Number of CRC patients: {num_crc_patients}")
print(crc_filtered_df["Gene Panel"].value_counts())
print(crc_filtered_df["Study ID"].value_counts())

crc_filtered_df
```
