## Use MSK-MET data to get frequency of metastasis to anatomical sites by primary cancer type

```python
import os
import csv
repo_dir = "/Users/divyakoyyalagunta/Desktop/Cornell_Research/Morris_Lab/met_history_prediction/"
os.chdir(repo_dir)
data_dir = os.path.join(repo_dir, "src/data/msk_met")
mks_met_csv_fn = os.path.join(data_dir, "msk_met_2021_clinical_data.tsv")
#msk_impact_csv_fn = os.path.join(data_dir, "msk_impact_2017_clinical_data.tsv")
```

```python
import pandas as pd
msk_met_df = pd.read_csv(mks_met_csv_fn, sep = '\t')
msk_met_df
```

```python
msk_met_df.columns
```

```python
cols_to_keep = ["Cancer Type"] + [x for x in list(msk_met_df.columns) if "Distant Mets:" in x]
msk_met_df = msk_met_df[cols_to_keep]
msk_met_df = msk_met_df.replace(to_replace = ['Yes','No'],value = [1,0])
msk_met_df
```

```python
summed_df = msk_met_df.groupby("Cancer Type").sum()
summed_df.columns = [x.replace("Distant Mets: ", "") for x in list(summed_df.columns)]
summed_df
```

```python
# Normalize
row_sums = summed_df.sum(axis=1)
print(row_sums)
freq_df = summed_df.div(row_sums, axis=0)
freq_df
```

```python
freq_df.to_csv(os.path.join(data_dir, "msk_met_freq_by_cancer_type.csv"))
```
