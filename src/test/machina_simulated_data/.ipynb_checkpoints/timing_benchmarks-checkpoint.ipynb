```python
import os
import pandas as pd

# Extract timing info from running MACHINA's pmh_ti on all m5 and m8 simulated data
repo_dir = "/Users/divyakoyyalagunta/Desktop/Cornell_Research/Morris_Lab/met_history_prediction/"
mach_time_fn = os.path.join(repo_dir, "src/test/machina_simulated_data/pmh_ti_timing_8cores_11232022.txt")

cols = ["seed", "site", "mig_type", "time"]
data = []
with open(mach_time_fn, "r") as f:
    for line in f:
        if "Runtime" in line:
            items = line.split()
            seed = int(items[3][:-1])
            mig_type = items[5][:-1]
            site = items[8][:-1]
            time = float(items[9])
            data.append([seed, site, mig_type, time])
machina_time_df = pd.DataFrame(data, columns=cols)
machina_time_df
```

```python
# Extract timing info from running gradient based method
gd_time_fn = os.path.join(repo_dir, "src/test/machina_simulated_data/predictions_wm10_wc5_wg0.5_stemp30_ftemp0.01_bs16_ssdecay_1core_11282022/results_wm10_wc5_wg0.5_stemp30_ftemp0.01_bs16_ssdecay_1core_11282022.txt")
gd_time_df = pd.read_csv(gd_time_fn).drop(columns=["mig", "comig", "seeding", "nll", "reg", "gen", "loss"])
gd_time_df
```

```python
# Get number of trees per site/mig_type/seed combo
tree_nums_df = gd_time_df.groupby(['site', 'mig_type', 'seed']).size().reset_index(name='counts')
tree_nums_df
```

```python
import fnmatch

# Get number of clusters (tree vertices to label) per site/mig_type/seed combo
sims_dir = os.path.join(repo_dir, "src/data/machina_sims")
data = []
for site in ["m5", "m8"]:
    for mig_type in ["mS", "M", "S", "R"]:
        site_mig_type_dir = os.path.join(sims_dir, site, mig_type)
        seeds = fnmatch.filter(os.listdir(site_mig_type_dir), 'clustering_observed_seed*.txt')
        seeds = [s.replace(".txt", "").replace("clustering_observed_seed", "") for s in seeds]
        for seed in seeds:
            fn = os.path.join(site_mig_type_dir, f"clustering_observed_seed{seed}.txt")
            num_lines = sum(1 for _ in open(fn))
            data.append([seed, site, mig_type, num_lines])
            
cols = ["seed", "site", "mig_type", "clusters"]
cluster_num_df = pd.DataFrame(data, columns=cols)
cluster_num_df
```

```python
# Get summed runtime for each site/mig_type/seed combo
gd_summed_time_df = gd_time_df.groupby(["site","mig_type", "seed"]).sum().drop(columns="tree_num").reset_index()
gd_summed_time_df
```

```python
# Collect timing data

cluster_nums = []
gd_avg_times = []
machina_avg_times = []
timing_data = []
for site in ["m5", "m8"]:
    for mig_type in ["mS", "M", "S", "R"]:
        site_mig_type_dir = os.path.join(sims_dir, site, mig_type)
        seeds = fnmatch.filter(os.listdir(site_mig_type_dir), 'clustering_observed_seed*.txt')
        seeds = [s.replace(".txt", "").replace("clustering_observed_seed", "") for s in seeds]
        for seed in seeds:
            # TODO: figure out why this isn't running on MACHINA
            if seed == '172' and site == "m8" and mig_type == "M":
                continue
            num_clusters = cluster_num_df[(cluster_num_df['site']==site) & (cluster_num_df['mig_type']==mig_type) & (cluster_num_df['seed']==seed)]['clusters'].item()
            num_trees = tree_nums_df[(tree_nums_df['site']==site) & (tree_nums_df['mig_type']==mig_type) & (tree_nums_df['seed']==int(seed))]['counts'].item()

            sum_gd_time = gd_summed_time_df[(gd_summed_time_df['site']==site) & (gd_summed_time_df['mig_type']==mig_type) & (gd_summed_time_df['seed']==int(seed))]['time'].item()
            sum_mach_time = machina_time_df[(machina_time_df['site']==site) & (machina_time_df['mig_type']==mig_type) & (machina_time_df['seed']==int(seed))]['time'].item()
            
            avg_gd_time = sum_gd_time/float(num_trees)
            avg_mach_time = sum_mach_time/float(num_trees)

            cluster_nums.append(num_clusters)
            gd_avg_times.append(avg_gd_time)
            machina_avg_times.append(avg_mach_time)
            timing_data.append([num_clusters, avg_gd_time, "Metient"])
            timing_data.append([num_clusters, avg_mach_time, "MACHINA"])

timing_df = pd.DataFrame(timing_data, columns=["Number of Tree Nodes", "CPU Wall-time (s)", "Method"])
timing_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
colors = ["#2496c8ff", "#ff915fff"]
sns.set_palette(sns.color_palette(colors))

with sns.plotting_context("notebook", font_scale=1.2):
    g = sns.relplot(
        data=timing_df, x="Number of Tree Nodes", y="CPU Wall-time (s)",
        height=5, aspect=1.1, dashes=False,
        hue="Method", style="Method", kind="line", ci='sd', legend=False,
    )

    (g.set_axis_labels("Number of Tree Nodes", "CPU Wall-time (s)")
      .set_titles("boop")
      .tight_layout(w_pad=0))
    
    plt.legend(title='', loc='upper left', labels=['Metient', 'Metient CI (1 SD)', "MACHINA", "MACHINA CI (1 SD)"], prop={'size': 10}, frameon=False)
    plt.show()

    fig = g.fig
    fig.savefig("timing_benchmarks.png", dpi=600) 
```
